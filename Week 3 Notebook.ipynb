### Gather Data from HTML Table from Wikipedia

Begin by using either `BeautifulSoup` or the `pandas.read_html` to read the data from a **URL**
into a `DataFrame` object and begin the ***prepocessing*** phase.

In this specific case, there is **one table** needed from the page for preprocessing, namely the one that 
have the data for each row / cell: 
* Postal Code
* Borough
* Neighborhood Name

The link can be found in <a href="https://en.wikipedia.org/wiki/List_of_neighbourhoods_in_Toronto" target="_link" title="Wikipedia Link">here</a>.

In *each* **table cell**, the general pattern is like this:
```html
{Postal-Code}
Borough ({Neighborhood-name})
```

Each row can be scraped by *iterating* through every `td` element in the table.
* **Postal Code** is obtained from the *first 3 elements* of the cell
* **Borough** is obtained by retrieving the text in the `span` element **before** the *parenthesis*
* **Neighborhood** is obtained by retrieving and formatting the text in `span` element after the `(`

````python
from collections import defaultdict
import re
url = "https://en.wikipedia.org/wiki/List_of_postal_codes_of_Canada:_M"

html_req = requests.get(url)
html_doc = html_req.text

soup     = BeautifulSoup(html_doc,'html.parser')
t        = soup.find_all('table')
table    = None
table_dict = defaultdict(list)
if len(t) > 0:
    table = t[0]
    
for cell in table.find_all('td'):
    span = cell.find('span')
    if span.text == "Not assigned":
        continue
    text         = cell.p.text
    PostalCode   = text[:3]
    Borough      = ""
    Neighborhood = ""
    split        = cell.span.text.split('(')
    if len(split) > 0:
        Borough = split[0]
    if len(split) > 1:
        Neighborhood = (((split[1].replace('(','').strip(')'))).replace(' /',',')).replace(')','').strip(' ')
    string       = "Postal Code: %s\nBorough: %s\nNeighborhood: %s\n" % (PostalCode,Borough,Neighborhood)
    table_dict['PostalCode'].append(PostalCode)
    table_dict['Borough'].append(Borough)
    table_dict['Neighborhood'].append(Neighborhood)
    
df = pd.DataFrame(data=table_dict,columns=['PostalCode','Borough','Neighborhood'])
df.dropna(inplace=True)
df['Borough']=df['Borough'].replace({'Downtown TorontoStn A PO Boxes25 The Esplanade':'Downtown Toronto Stn A',
                                             'East TorontoBusiness reply mail Processing Centre969 Eastern':'East Toronto Business',
                                             'EtobicokeNorthwest':'Etobicoke Northwest','East YorkEast Toronto':'East York/East Toronto',
                                             'MississaugaCanada Post Gateway Processing Centre':'Mississauga'})
df.head()
````

Since each cell was a combination of Postal Code, Borough, and Neighborhoods in one *text-block*,
I simply extracted each feature by separating the `span` elements and splitting words before
and the `()` to extract neighborhood names.

I follow by reencoding the `Borough` names into shorter more concise names using the `replace` method.

## Gather Venue Data from Foursquare API

Use the credentials from **Foursquare API Account** to gather 
the *nearby venues* for **each neighborhood**.

````python
def gatherNearbyVenues(names,latitudes,longitudes,client_id,client_secret,radius=500,version='20180605',limit=100):
    data_map = defaultdict(list)
    
    for name, latitude, longitude in zip(names,latitudes,longitudes):
        # create the API request URL
        url = 'https://api.foursquare.com/v2/venues/explore?&client_id={}&client_secret={}&v={}&ll={},{}&radius={}&limit={}'.format(
            client_id, 
            client_secret, 
            version, 
            latitude, 
            longitude,
            radius, 
            limit)
        json = requests.get(url).json()
        if 'response' in json and 'groups' in json['response']:
            if len(json['response']['groups']) > 0 and 'items' in json['response']['groups'][0]:
                results = json['response']['groups'][0]['items']
                for v in results:
                    data_map['Neighborhood'].append(name)
                    data_map['Neighborhood Latiude'].append(latitude)
                    data_map['Neighborhood Longitude'].append(longitude)
                    data_map['Venue'].append(v['venue']['name'])
                    data_map['Venue Latitude'].append(v['venue']['location']['lat'])
                    data_map['Venue Longitude'].append(v['venue']['location']['lng'])
                    data_map['Venue Category'].append(v['venue']['categories'][0]['name'])
        df = pd.DataFrame(data = data_map)
    return df
````

### Aggregate Toronto Data For Clustering

## K Means Cluster Analysis
Begin the K Means cluster analysis by grouping the averaging the counts of
every **venue** per **neighborhood**.

````python
num_clusters       = 5
foursquare_grouped = foursquare_onehot.groupby('Neighborhood').mean().reset_index()
foursquare_grouped.head()
````

From here we can initialize a **K Means Cluster** with `k` set to `6`
and fit our **model** to the dataset

````python
foursquare_cluster_df = foursquare_grouped.drop('Neighborhood',axis=1)

## Fit KMeans cluster to FourSquare Dataset of Averages
kmean = KMeans(n_clusters=num_clusters,random_state=0).fit(foursquare_cluster_df)

#Verify Label
kmean.labels_[0:10]
````

### Retrieve Ranking of Each Neighborhood's Most Popular Venues
We will also want to see the most popular venues so we can visually see a pattern 
in what are the top venues of each cluster and how they might be related

````python
def return_most_common_venues(row, num_top_venues):
    row_categories = row.iloc[1:]
    row_categories_sorted = row_categories.sort_values(ascending=False)
    
    return row_categories_sorted.index.values[0:num_top_venues]

num_top = 10
````

````python
indicators = ['st', 'nd', 'rd']

# create columns according to number of top venues
columns = ['Neighborhood']
for ind in np.arange(num_top):
    try:
        columns.append('{}{} Most Common Venue'.format(ind+1, indicators[ind]))
    except:
        columns.append('{}th Most Common Venue'.format(ind+1))

# create a new dataframe
neighborhoods_venues_sorted = pd.DataFrame(columns=columns)
neighborhoods_venues_sorted['Neighborhood'] = foursquare_grouped['Neighborhood']

for ind in np.arange(foursquare_grouped.shape[0]):
    neighborhoods_venues_sorted.iloc[ind, 1:] = return_most_common_venues(foursquare_grouped.iloc[ind, :], num_top)

neighborhoods_venues_sorted.head()
````

````python
# Add a Column to Dataset that includes the generated labels for each neighborhood from KMeans
neighborhoods_venues_sorted.insert(0,'Cluster Label',kmean.labels_)

merged_data = toronto_df
## Join Data By Neighborhood
merged_data = merged_data.join(neighborhoods_venues_sorted.set_index('Neighborhood'),on='Neighborhood')

````
